```python
numbers = []

for i in range(5):
    num = float(input("Enter a number: "))
    numbers.append(num)


total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)


print("Numbers:", numbers)
print("Sum:", total)
print("Average:", average)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
words_list = []

for i in range(5):
    word = input("Enter a word: ")
    words_list.append(word)

words = tuple(words_list)


print("First word:", words[0])
print("Last word:", words[-1])
print("Total number of words:", len(words))


search_word = input("Enter a word to search: ")

if search_word in words:
    print("Word found in tuple.")
else:
    print("Word not found in tuple.")
```

```python
students = {}


for i in range(3):
    name = input("Enter student name: ")
    score = float(input("Enter student score: "))
    students[name] = score


print("Students and their scores:")
for name, score in students.items():
    print(name, ":", score)


highest_student = max(students, key=students.get)

print("Student with highest score:", highest_student)
print("Highest score:", students[highest_student])
```

```python
list1 = list(map(int, input("Enter numbers for list 1 separated by spaces: ").split()))
list2 = list(map(int, input("Enter numbers for list 2 separated by spaces: ").split()))


set1 = set(list1)
set2 = set(list2)


print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
print("Difference (set1 - set2):", set1.difference(set2))
```

```python
names = []

for i in range(5):
    name = input("Enter a name: ")
    names.append(name)


names_tuple = tuple(names)
print("Tuple:", names_tuple)


name_dict = {}

for name in names:
    name_dict[name] = len(name)

print("Dictionary (name:length):", name_dict)


length_set = set(name_dict.values())

print("Set of unique name lengths:", length_set)
```
