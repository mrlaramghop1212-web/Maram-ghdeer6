```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    result = num1 / num2

except (ZeroDivisionError, ValueError):
    print("Error: Division by zero or invalid input.")

else:
    print("Result:", result)
```

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self._age = age

    def update_age(self, new_age):
        self._age = new_age

    def display(self):
        print("Name:", self.name)
        print("Age:", self._age)



person1 = Person("Maram", 18)


person1.update_age(20)


person1.display()
```

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


class Manager(Employee):

    def add_bonus(self, bonus):
        self.salary += bonus

    def display_info(self):
        print("Name:", self.name)
        print("Total Salary:", self.salary)



manager1 = Manager("Maram", 10000)


manager1.add_bonus(1000)


manager1.display_info()
```

```python
class Student:

    def __init__(self):
        self.name = input("Enter name: ")

        try:
            age_input = input("Enter age: ")
            self.age = int(age_input)

            if self.age < 0:
                print("Error: Age must not be negative.")
                self.age = None

        except ValueError:
            print("Error: Age must be a number.")
            self.age = None

    def display(self):
        if self.age is not None:
            print("Name:", self.name)
            print("Age:", self.age)



student1 = Student()
student1.display()
```
