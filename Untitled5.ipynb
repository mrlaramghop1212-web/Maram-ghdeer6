```python
import os

base_dir = "lab_data"
sub_folders = ["input", "output", "logs"]

if not os.path.exists(base_dir):
    os.makedirs(base_dir)
    print(f"Folder '{base_dir}' created.")
else:
    print(f"Folder '{base_dir}' already exists.")

for folder in sub_folders:
    path = os.path.join(base_dir, folder)
    if not os.path.exists(path):
        os.makedirs(path)

print("\nContents of lab_data:")
for item in os.listdir(base_dir):
    item_path = os.path.join(base_dir, item)
    if os.path.isdir(item_path):
        print(f"{item} is a Directory")
    else:
        print(f"{item} is a File")

old_path = os.path.join(base_dir, "logs")
new_path = os.path.join(base_dir, "history")

if os.path.exists(old_path):
    os.rename(old_path, new_path)
    print("\nRenamed 'logs' to 'history'.")
```

```python
input_path = os.path.join("lab_data", "input", "notes.txt")

with open(input_path, "w") as f:
    f.write("Line 1: Welcome to Lab 8\n")
    f.write("Line 2: Python is great for OS tasks\n")
    f.write("Line 3: Working with directories is easy\n")
    f.write("Line 4: This is the fourth line\n")
    f.write("Line 5: End of the notes file\n")

with open(input_path, "r") as f:
    lines = f.readlines()
    print(f"\nTotal number of lines in notes.txt: {len(lines)}")

output_path = os.path.join("lab_data", "output", "summary.txt")

with open(output_path, "w") as f_out:
    f_out.writelines(lines)

print("Contents copied to summary.txt.")
```
